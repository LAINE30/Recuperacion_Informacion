### Paso 1: Calcular la matriz TF-IDF para el corpus `data/01_corpus_turismo_500.txt`

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 1. Cargar el corpus
with open('C:\\Users\\luis3\\Downloads\\RI26A\\Recuperacion_Informacion\\data\\01_corpus_turismo_500.txt', 'r', encoding='utf-8') as f:
    corpus_turismo = f.readlines()

# 2. Inicializar el vectorizador (puedes añadir stop_words='spanish')
vectorizer = TfidfVectorizer()

# 3. Calcular la matriz TF-IDF
tfidf_matrix_turismo = vectorizer.fit_transform(corpus_turismo)

# Opcional: Ver las primeras filas en un DataFrame para inspección
nombres_terminos = vectorizer.get_feature_names_out()
df_turismo = pd.DataFrame(tfidf_matrix_turismo.toarray(), columns=nombres_terminos)
print(df_turismo.head())
```

### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
# Importar bibliotecas necesarias
import os
import re
from sklearn.metrics.pairwise import cosine_similarity
# Listar los archivos en el directorio
path = r'C:\\Users\\luis3\\Downloads\\RI26A\\Recuperacion_Informacion\\data\\gutenberg_top100\\'
files = os.listdir(path)
print(files)
len(files)
```

```python
# Función para limpiar el texto 
# (eliminar encabezados, pies, convertir a minúsculas, eliminar caracteres no alfabéticos)
def limpiar_texto(texto):
    # Eliminar encabezados y pies de Gutenberg
    inicio = texto.find("*** START OF")
    fin = texto.find("*** END OF")
    if inicio != -1 and fin != -1:
        texto = texto[inicio:fin]

    # Minúsculas
    texto = texto.lower()

    # Eliminar todo lo que no sea letras (quita números, chino, símbolos)
    texto = re.sub(r'[^a-z\s]', ' ', texto)

    # Quitar espacios extra
    texto = re.sub(r'\s+', ' ', texto).strip()

    return texto
```

```python
# Limpiar y cargar el corpus de Gutenberg
corpus = []
# Iterar sobre los archivos, leer su contenido, limpiarlo y añadirlo al corpus
for filename in files:
    file_path = os.path.join(path, filename)

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            texto = f.read()
    except:
        with open(file_path, 'r', encoding='latin-1') as f:
            texto = f.read()
    # Limpiar el texto
    texto_limpio = limpiar_texto(texto)
    corpus.append(texto_limpio)
# Verificar que se han cargado los documentos
print(f'Documentos cargados: {len(corpus)}')

# Ver ejemplo
print("\nEjemplo de texto limpio:\n")
print(corpus[0][:500])
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

```python
# Inicializar el vectorizador con parámetros para reducir la dimensión y eliminar palabras comunes
vectorizer = TfidfVectorizer(
    stop_words='english',
    max_features=10000,   # reduce dimensión
    max_df=0.8,
    min_df=5
)

tfidf_matrix = vectorizer.fit_transform(corpus)

print("TF-IDF calculado")
```

```python
# Convertir la matriz TF-IDF a un DataFrame para inspección
feature_names = vectorizer.get_feature_names_out()

df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names
)

print(df.head())
print(f"\nDimensión del DataFrame: {df.shape}")
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
# Función de búsqueda que recibe una consulta, 
# vectoriza la consulta, 
# calcula similitud coseno con los documentos y devuelve los resultados ordenados por relevancia
def buscar_palabra(query, top_k=5):
    # Vectorizar la consulta
    query_vec = vectorizer.transform([query])

    # Calcular similitud coseno
    similitudes = cosine_similarity(query_vec, tfidf_matrix).flatten()

    # Obtener índices ordenados (mayor a menor)
    indices = similitudes.argsort()[::-1][:top_k]

    # Mostrar resultados
    resultados = []
    for i in indices:
        resultados.append({
            "documento": files[i],   # o nombres_docs si usaste esa variable
            "score": similitudes[i]
        })

    return pd.DataFrame(resultados)
```

```python
# Ejemplo de búsqueda
buscar_palabra("zarathustra")
```
